# Example of a CNT system.

 Example of using trajectory-kit for analysing ionic permeation of K<sup>+</sup> into the CNT.

# 1. Basic Setup

```python
from trajectory_kit import sim

import json 

json_path = "" # <-- change this to your own path

with open(json_path, "r") as f:
    paths = json.load(f)

cnt_pdb_path = paths["pdb"]
cnt_psf_path = paths["psf"]
cnt_dcd_path = paths["dcd"]

s = sim(
    typing=cnt_pdb_path,
    topology=cnt_psf_path, 
    trajectory=cnt_dcd_path,
)

s.print_info()
```

# 2. Trajectory Analysis - Positions

```python
TYPE_QUERY = {
    "atom_name": {"K"}
}

TOPO_QUERY = {
    "atom_type": {"POTD"}
}

TRAJ_QUERY = {
    "frame_interval": (0, None, 1)
}

out11 = s.positions(TYPE_Q={}, TOPO_Q=TOPO_QUERY, TRAJ_Q=TRAJ_QUERY)
out12 = s.positions(TYPE_Q=TYPE_QUERY, TOPO_Q={}, TRAJ_Q=TRAJ_QUERY)

for key in out11.keys():
    print(f"{key} ")

positions, meta = out11["payload"]

print("Positions shape:", positions.shape)
```

## 3. Trajectory Analysis - Fetch

```python
TOPO_QUERY = {
    "atom_type": {"CD2R6A"}
}

TRAJ_QUERY = {
    "frame_interval": (0, None, 1)
}

out31 = s.fetch(TYPE_Q={}, TOPO_Q=TOPO_QUERY, TRAJ_Q=TRAJ_QUERY, TYPE_R="positions", TOPO_R="charges", TRAJ_R="global_ids")
out31
```
