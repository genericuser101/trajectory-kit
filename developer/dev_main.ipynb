```python
from trajectory_kit import sim
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
def find_first_with_extension(folder, ext):
    for f in sorted(os.listdir(folder)):
        if f.lower().endswith(ext.lower()):
            path = os.path.join(folder, f)
            print(f"Found {path}")
            return path
    return None
```
